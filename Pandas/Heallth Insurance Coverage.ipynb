# <div class="alert alert-success">Comparison between medicare enrollment and uninsured rate</div>


```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = [14.6, 19.9, 16.9, 17.5, 18.5, 15.9, 9.1]
y = [10.1, 14.9, 10.8, 9.5, 8.6, 8.1, 6]

# Set the width of the bars
bar_width = 0.35

# Generate the x-axis labels
labels = np.arange(len(x))

# Create a bar chart with x and y data
plt.bar(labels, x, bar_width, label='uninsured rate')
plt.bar(labels + bar_width, y, bar_width, label='medicare enrollment')

# Add labels and title
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Comparison of medicare enrollmenr and uninsured rate')

# Add legend
plt.legend()

# Show the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create figure and axis objects
fig, ax = plt.subplots(figsize=(10, 8))

# Create stacked area plot for x and y
ax.stackplot(range(len(x)), x, y, labels=['x', 'y'])

# Add axis labels and legend
ax.set_xlabel('Data points')
ax.set_ylabel('Values')
plt.legend(loc='upper left')

# Show the plot
plt.show()

```
